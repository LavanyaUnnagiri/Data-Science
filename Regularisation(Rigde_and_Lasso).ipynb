### Regularization

- Tuning Technique
- to solve the problem of overfitting
- reduce the errors by fitting the function approriately on the training data and avoid the overfitting
- slighltly do modifications to the learning algorithm
- improves the model perfomance on the unseen data
### 3 Techniques

- Ridge
- Lasso
- Elastic net
### overfitting
- training data error is less
- test data error is more #### Ridge regression(L2 Regularization)
- costfunction+(lamba*(slope)^2)

```python
### Ridge Regression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
df=load_boston()
df
```

```python
df.keys()
```

```python
boston=pd.DataFrame(df.data)
```

```python
boston.columns=df.feature_names
```

```python
boston.head()
```

```python
boston["Target"]=df.target
boston
```

```python
boston.isnull().sum()
```

```python
x=boston[boston.columns[:-1]]
x
```

```python
y=boston["Target"]
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scale=scale.fit_transform(x)
```

```python
x_scale
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,train_size=0.7,random_state=42)
model=Ridge(alpha=1)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_test,y_test)*100
```

```python
model.score(x_train,y_train)*100
```

### Lasso Regression(L1 Regularization)

- Identical to the ridge(reduces the overfitting)
- costfuncion+lambda*|slope|--->formula
- removes unwanted features(magintudes of the unwanted features will be zero)

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer_df=pd.DataFrame(cancer.data)
```

```python
cancer_df.columns=cancer.feature_names
cancer_df.head()
```

```python
cancer_df["target"]=cancer.target
```

```python
cancer_df
```

```python
x=cancer_df[cancer_df.columns[:-1]]
y=cancer["target"]
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=1)
model=Lasso(alpha=1)
model.fit(x_train,y_train)
```

```python
coeff_used=np.sum(model.coef_!=0)
coeff_used
```

```python
#try to change the alpha values check your testing and training perfomance
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=100)
model=Lasso(alpha=0.0001)
model.fit(x_train,y_train)
```

```python
y_pedict=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
